**Programas numéricos**

Existen, diversos tipos de números en Python, pero podemos ver que, dentro de los tipos de datos, existen 2, los enteros (int) y los flotantes (float)
Dentro de este concepto de programas numéricos, existe la enumeración exhaustiva.

**Enumeración exhaustiva**
* También conocido como "adivina y verifica"
* Las computadoras actuales son muy muy rápidas
* Puede ser de los algoritmos más importantes

```python
numero = int(input("Ingresa un número entero: "))

if type(numero) != int:
    print("Se ha producido un error, no hay un número entero")
else:
    respuesta = 0
    while respuesta**2 < numero:
        print(respuesta)
        respuesta+=1

    if respuesta**2  == numero:
        print(f'La raiz cuadrada de {numero} es {respuesta}')
    else:
        print(f'El objetivo no tiene raiz cuadradada exacta')
```

**Aproximación de soluciones**

* En computación no se puede ser preciso y eficiente a la vez

* Se debe buscar que es lo mas adecuado para la solución

```python
#En el siguiente código, se pueden encontrar raíces de números aproximados
numero = int(input("Escoger un número para calcular la raiz"))
#epsilon, medición de la presición
epsilon = 0.01
#número máximo de iteraciones
paso = epsilon**2
contador = 0.0


#Crear un bucle while con el valor absoluto
#Mientras el valor absoluto del contador menos el número se acerque al epsilon
while abs(contador**2 - numero) >= epsilon and contador <= numero:
    print(abs(contador**2), contador)
    contador += paso

if abs(contador**2 - numero) >= epsilon:
    print("No hay solución en los límites dados")
else:
    print(f'La raiz cuadrada de {numero} es {contador}')
```

**Búsqueda binaria**

La búsqueda binaria es un algoritmo para ordenar una serie de números, se basa en comparar pares y hacer la comparación de cada elemento.

```python
objetivo = int(input("Escoge un número"))
epsilon = 0.01
bajo = 0.0
alto = max(1.0, objetivo)
respuesta = (alto + bajo) / 2

#Se divide entre 2 el espacio de búsqueda y se redefine que es alto y que es bajo
while abs(respuesta ** 2 - objetivo) >= epsilon:
    print(f'bajo = {bajo}, alto = {alto}')
    if respuesta ** 2 <= objetivo:
        bajo = respuesta
    else:
        alto = respuesta

    respuesta = (alto + bajo) / 2

print(f'La raiz cuadrada de {objetivo} es {respuesta}')
```

**Cálculo del Factorial**

```python
def factorial(n):
    '''
    Esto es una función para calcular el factorial en Python

    n int > 0

    returns n!

    '''

    if n == 1:
        return 1
    
    return n * factorial(n - 1)
    

n = int(input("Escribe un número :"))
print(f'El factorial del numero {n} es {factorial(n)}')
```

```python
def aplicar_operaciones(num):
    operaciones = [abs, float]

    resultado = []

    for operacion in operaciones:
        resultado.append(operacion(num))
    
    return resultado

aplicar_operaciones(-2)
```

```python
def primera_letra(lista_de_palabras):
    #Docstring
    """Esta función recibe una lista de palabras y devuelve la primera letra de cada una."""
    primeras_letras = []

    for palabra in lista_de_palabras:
        assert type(palabra) == str, f'{palabra} no es string'
        assert len(palabra) > 0, "No se permiten vacíos"

        primeras_letras.append(palabra[0])
    return primeras_letras

lista_de_palabras = ['Hola', 'Mundo', 'Celular', 'Palabra', 'Rojo']
print(primera_letra(lista_de_palabras))
```
